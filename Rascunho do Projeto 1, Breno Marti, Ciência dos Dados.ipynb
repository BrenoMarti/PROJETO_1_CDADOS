###  Primeira análise

Os dados plotados em forma de pizza mostram que, nos anos de 1970, 1980 e 2000, os EUA representam uma parcela gigantesca dos gastos militares absolutos da OTAN. O que indica que parte da fala de Trump se verifica verdadeiramente, porém, antes de se fazer qualquer conclusão a respeito do assunto, é necessária uma análise mais elaborada dos dados, levando em conta uma visualização mais contínua da variação na participação dos países dentro dos gastos da OTAN e compará-los com outros países, a fim de verificar a hipotese anteriormente mostrada.

```python
# lista adaptada para cálculo da participação nos gastos da OTAN pelos EUA ao longo dos anos sem os países sem dados nas
# tabelas e somente com os países de 1970, sem a inclusão de novos membros

paises_otan_EUA = ['Belgium',
                    'Canada', 
                    'Denmark', 
                    'France',  
                    'Italy', 
                    'Luxembourg',  
                    'Netherlands', 
                    'Norway', 
                    'Portugal',  
                    'United Kingdom', 
                    'Greece', 
                    'Turkey', 
                    'Germany',
                    'United States',]
```

```python
OTAN_M = GDP_militar[paises_otan_EUA]

OTAN_E = GDP_total[paises_otan_EUA]
```

```python
i=0
j=0
particiEUA=[]
total_ano=0
soma=0

#While que calcula todos os anos
while i<58:
    
    #estabelesce o valor inicial de cada ano i 
    total_ano=0
    j=0
    
    #Calcula a soma de todas as partes da OTAN no ano i
    while j<len(paises_otan_EUA):
        
        #mecanismo de cáculdo do valor total gasto em exercito pela otan - EUA no ano i
        soma = (OTAN_M.iloc[i,j]) * OTAN_E.iloc[i,j]
        total_ano += soma
        j+=1
        
    #Divide a parcela dos EUA pela soma das outras partes para o ano i, multiplicando por 100 para deixar em %
    parte_EUA = OTAN_M.iloc[i,len(paises_otan_EUA)-1] * OTAN_E.iloc[i,len(paises_otan_EUA)-1]
    particiEUA.append((parte_EUA/total_ano)*100)
    
    #progride para o próximo ano i
    i+=1
    
```
