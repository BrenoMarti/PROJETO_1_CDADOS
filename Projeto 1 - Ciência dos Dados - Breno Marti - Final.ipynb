Durante os anos e ao escalar da guerra fria, a OTAN ganhou mais membros, totalizando quase toda a Europa Ocidental. Hoje em dia, a organização demonstra uma hegemonia completa no cenário militar mundial, concentrando valores próximos de 70% de todo o poderio militar do globo.

### O que se espera dos dados para uma confirmação:

> A participação dos Estados Unidos nos gastos militares dentro da OTAN deve representar uma parcela majoritaria.

Cada país terá uma série produzida por meio do código, em que a sua contribuição seja dividida pelo valor de todas as outras mais a sua própria, para que possa-se calcular a evolução da porcentagem de sua contribução ao longos dos anos de 1970 até 2018.

#### Observações:

O valor total incluirá todos os membros da OTAN no ano de 1970, com exceção da Islândia, cujos dados não estão presentes no dataframe do World Bank.
```python
# lista adaptada para cálculo da participação nos gastos da OTAN pelos EUA ao longo dos anos sem os países sem dados nas
# tabelas e somente com os países de 1970, sem a inclusão de novos membros

paises_otan_EUA = ['Belgium',
                    'Canada', 
                    'Denmark', 
                    'France',  
                    'Italy', 
                    'Luxembourg',  
                    'Netherlands', 
                    'Norway', 
                    'Portugal',  
                    'United Kingdom', 
                    'Greece', 
                    'Turkey', 
                    'Germany',
                    'United States',]
```

```python
OTAN_M = GDP_militar[paises_otan_EUA]

OTAN_E = GDP_total[paises_otan_EUA]
```

```python
i=0
j=0
particiEUA=[]
total_ano=0
soma=0

#While que calcula todos os anos
while i<58:
    
    #estabelesce o valor inicial de cada ano i 
    total_ano=0
    j=0
    
    #Calcula a soma de todas as partes da OTAN no ano i
    while j<len(paises_otan_EUA):
        
        #mecanismo de cáculdo do valor total gasto em exercito pela otan - EUA no ano i
        soma = (OTAN_M.iloc[i,j]) * OTAN_E.iloc[i,j]
        total_ano += soma
        j+=1
        
    #Divide a parcela dos EUA pela soma das outras partes para o ano i, multiplicando por 100 para deixar em %
    parte_EUA = OTAN_M.iloc[i,len(paises_otan_EUA)-1] * OTAN_E.iloc[i,len(paises_otan_EUA)-1]
    particiEUA.append((parte_EUA/total_ano)*100)
    
    #progride para o próximo ano i
    i+=1
```
